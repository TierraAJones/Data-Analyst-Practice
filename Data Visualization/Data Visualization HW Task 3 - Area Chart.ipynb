Create a Stacked Area Chart of Four Engine Categories

Create a stacked area chart using all four engine type categories on the chart. They are Gas, Petrol, Diesel, and Other.

Think about the order in which you'll display the categories. Consider the colors, which would fit your graph. 

Finally, think about a setting in which you would be required to present all four categories from the data. What would be the interpretation of your newly created chart?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
```

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ["#011638", "#7e2987", "#ef2026", "#FFC000"]
labels = ["Diesel", "Petrol", "Gas", "Other"]
sns.set_style("white")
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              df_fuel_engine_types["Other"],
              colors = colors,
              edgecolor = 'none')
plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```
